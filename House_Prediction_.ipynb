Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing(as_frame=True)
```

```python
print(california)
```

```python
# Features (X) and target (y)
X = california.data  # DataFrame with 8 features
y = california.target  # Median house value (in $100,000s)
```

```python
#Loading the dataset to pandas dataframe
california_data = pd.DataFrame(X,columns=california.feature_names)
```

```python
california_data.head()
```

```python
#add the target (price) column to the DataFrame
california_data['MedHouseVal'] = y  # Add target column
```

```python
print(california_data.head())
```

```python
# checking the number of rows and Columns in the data frame
california_data.shape
```

```python
# check for missing values
california_data.isnull().sum()
```

```python
# statistical measures of the dataset
california_data.describe()
```

Understanding the correlation between various features in the dataset

Positive Correlation

Negative Correlation

```python
correlation = california_data.corr()
```

```python
# constructing a heatmap to nderstand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and Target

```python
print(X)
```

```python
print(y)
```

Splitting the data into Training data and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```

Prediction on Test Data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```
